```python
import matplotlib.pyplot as plt
import json
from pathlib import Path
```

```python
def load_group(meta_paths):
    meta_json_paths = [[f for f in (Path(x)/".logs").iterdir() if f.suffix == ".json"][0] for x in meta_paths]
    metas = [json.load(open(pth, "r")) for pth in meta_json_paths]
    durations = [meta["total"]["duration"] for meta in metas]
    print(durations)
    return durations
```
