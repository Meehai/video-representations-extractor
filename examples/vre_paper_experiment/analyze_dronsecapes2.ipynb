```python
def read_all_get_max(pth: Path) -> float:
    pth = Path(pth)
    all_metas = [x for x in (Path(pth)/".logs").iterdir() if x.suffix == ".json"]
    datas = [json.load(open(meta_path, "r")) for meta_path in all_metas]
    all_durations = [x["total"]["duration"] for x in datas]
    print(f"{pth.name} has {len(all_metas)} metadatas. Max: {max(all_durations)}. {all_durations=}.")
    return max(all_durations)    

data = read_all_get_max("data_video_540_960/")
data2 = read_all_get_max("data_video_540_960_parallel_2/")
data4 = read_all_get_max("data_video_540_960_parallel_4/")
data6 = read_all_get_max("data_video_540_960_parallel_6/")
data8 = read_all_get_max("data_video_540_960_parallel_8/")
res = [data, data2, data4, data6, data8]
```

```python
xs = [1, 2, 4, 6, 8]
res2 = [1/(xs[i]*res[i]/res[0]) for i in range(len(xs))]
print(res2)
plt.plot(xs, res)
plt.xticks(xs)
plt.title("Total duration for Dronescapes 2 config (N=100 frames)")
plt.ylabel("Total duration (s)")
plt.xlabel("# GPUs used")
plt.show()
# plt.plot(xs, res2)
# plt.show()
```
